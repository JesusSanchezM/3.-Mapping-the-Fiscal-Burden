```python
import pandas as pd

# Ruta del archivo
ruta = "/Users/jesussanchez/Downloads/Censo2020_CA_pue_csv/Personas21.CSV"

# Cargar únicamente columnas necesarias
df = pd.read_csv(
    ruta,
    usecols=["MUN", "ALFABET", "FACTOR"],
    encoding="latin1",
    low_memory=False
)

# Revisar valores únicos de alfabetización
print(df["ALFABET"].value_counts(dropna=False))

# Asegurar que FACTOR sea numérico
df["FACTOR"] = pd.to_numeric(df["FACTOR"], errors="coerce")

# Tabulado ponderado por FACTOR
tabulado = (
    df
    .groupby(["MUN", "ALFABET"], dropna=False)["FACTOR"]
    .sum()
    .reset_index()
)

# Convertir a formato ancho
tabla_final = (
    tabulado
    .pivot(index="MUN", columns="ALFABET", values="FACTOR")
    .fillna(0)
)

# Opcional: redondear
tabla_final = tabla_final.round(0)

# Mostrar resultado
print(tabla_final)

# Guardar a CSV
tabla_final.to_csv("alfabetizacion_por_municipio_puebla.csv")

print("Archivo guardado: alfabetizacion_por_municipio.csv")


# Exportar a Excel
archivo_salida = "alfabetizacion_por_municipio_puebla.xlsx"

with pd.ExcelWriter(archivo_salida, engine="openpyxl") as writer:
    tabla_final.to_excel(writer, sheet_name="Tabulado")

print(f"Archivo Excel guardado como: {archivo_salida}")
```
